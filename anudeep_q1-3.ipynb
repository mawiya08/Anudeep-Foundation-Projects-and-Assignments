```python
"""A company wants to identify whether an employee is eligible for a bonus based on attendance percentage and
performance rating."""
employee=int(input("Enter No of employees"))
for i in range (employee):
  attendance=float(input(f"Enter attendance for {i+1} in %:"))
  rating=float(input(f"Enter performance rating for {i+1} in %:"))
  if (attendance>=90.0) and (rating>=4.0):
    print ("bonus")
  else:
    print('no bonus')
```

```python
""" Create a program to check whether a given username and password match predefined credentials"""
default_username="admin"
default_password="Qwerty12"
username=input("Enter user name:")
password=input("Enter password")
if (default_username==username) and (default_password==password):
  print("Correct Credentials")
else:
  print("Incorrect Credentials")
```

```python
"""A traffic monitoring system needs to determine whether a vehicle exceeded the speed limit"""
road_speed_limit=float(input("Enter road speed limit in km/hr: "))
vehicle_speed=float(input("Speed of moving vehicle: "))
if (0< vehicle_speed<= road_speed_limit):
  print("within speed limit")
elif( vehicle_speed>road_speed_limit):
  print("overspeeding")
```
