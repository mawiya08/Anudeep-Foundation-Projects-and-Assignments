```python
" a bank wants to calculate simple interest using principle, rate, and time "
principle=float(input("Enter Principle value: "))
rate=float(input("Enter Rate: "))
time=float(input("Enter time in years: "))
si=principle*rate*time
print(f"SI: {si}")
```

```python
"A movie theatre wants to assign ticket pricing based on age categories such as child, adult, and senior citizen."
age=int(input("Enter age: "))
if age<0:
  print("Invalid")
#for children
elif age<=12:
  print("Child")
  print("$5.00")
#for adults
elif age<=64:
  print("Adult")
  print("$15.00")
#for Senior Citizen
else:
  print("Senior Citizen")
  print("$10.00")
```
